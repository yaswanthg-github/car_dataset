### Data Cleaning

```python
# Creating 2 columns, one for the Car model name and the other for the company of the car 
#Splitting company name from CarName column
new = car["CarName"].str.split(" ", n = 1, expand = True)
  
# making separate first name column from new data frame
car["Company"]= new[0]
  
# making separate last name column from new data frame
car["Model"]= new[1]
  
# Dropping old Name columns
car.drop(columns =["CarName"], inplace = True)
```

```python
car.head()
```

```python
# Unique values in Company column
car['Company'] = car['Company'].str.capitalize()
car['Company'].unique()
```

```python
# There were a couple of spelling mistakes in the Company column, for example
# maxda(mazda)
# The Company column is updated with the correct company names
car.Company.replace('Maxda','Mazda',inplace=True)
car.Company.replace('Porcshce','Porsche',inplace=True)
car.Company.replace('Toyouta','Toyota',inplace=True)
car.Company.replace('Vokswagen','Volkswagen',inplace=True)
car.Company.replace('Vw','Volkswagen',inplace=True)

# Resulting Company column after correction
car['Company'].unique()
```

```python
# Segregating categorical data from car into new variable except Company and Model

cat_data = [feature for feature in car.columns if car[feature].dtypes=='O' and feature!='Company' and feature!='Model']
```

```python
car[cat_data]
```

```python
#check for unique values

for feature in cat_data:
    print(feature, car[feature].unique())
```

### EDA on Categorical Data

```python
# Categorical features
cat_data
```

```python
# Plotting a frequency table for cat_data
def frequency(x):   #helper function
    sns.barplot(x=car[x].value_counts().index, y=car[x].value_counts())
    plt.xlabel(x)
    plt.ylabel("Frequency")
    plt.title("Frequency table for %s" %x)
    plt.show()
    
for features in cat_data:
    frequency(features)
```

```python
# Frequency table for company
plt.figure(figsize=(20,6))
frequency('Company')
```

```python
#countplot for categorical data

plt.figure(figsize=(30,30))
ax = plt.subplots()

def cat_helper(x):    #helper function for countplot
    plt.xlabel(x)
    plt.bar(car[x], car['price'])
    plt.ylabel("Price of car")
    plt.show()
    

for features in cat_data:
    cat_helper(features)
    
plt.tight_layout()
```
